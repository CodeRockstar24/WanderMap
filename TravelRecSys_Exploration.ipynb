```python
import heapq
import numpy as np
from sklearn.neighbors import BallTree

# Build hashmap: cluster label to list of (index, row) tuples
cluster_to_dest = {}
for idx, row in datarecords.iterrows():
    cluster_to_dest.setdefault(row['Cluster_Label'], []).append((idx, row))

# Convert Latitude/Longitude to radians for BallTree haversine metric
coords = np.radians(datarecords[['Latitude', 'Longitude']].values)

# Build BallTree for location queries
balltree = BallTree(coords, leaf_size=30, metric='haversine')

def recommend_travel(user_destination, top_n=5, use_location=True, filter_type=None,
                     weight_popularity=0.7, weight_proximity=0.3):
    user_rows = datarecords[datarecords['Destination_Name'] == user_destination]
    if user_rows.empty:
        return "Destination '" + user_destination + "' not found."

    user_row = user_rows.iloc[0]
    user_cluster = user_row['Cluster_Label']

    user_lat = float(user_row['Latitude'])
    user_lon = float(user_row['Longitude'])
    user_coord = np.radians(np.array([user_lat, user_lon])).reshape(1, -1)

    # Filter cluster destinations by type if specified, exclude user's destination
    cluster_dests = [item for item in cluster_to_dest[user_cluster] if item[1]['Destination_Name'] != user_destination]
    if filter_type:
        cluster_dests = [item for item in cluster_dests if item[1]['Type'] == filter_type]

    if not cluster_dests:
        return f"No other destinations in cluster {user_cluster} with type '{filter_type}' to recommend." if filter_type else f"No other destinations in cluster {user_cluster} to recommend."

    # Heapify by negative popularity, using index as tie-breaker
    heap = [(-d[1]['Popularity_Score'], d[0], d[1].to_dict()) for d in cluster_dests]
    heapq.heapify(heap)
    top_cluster_recs = [heapq.heappop(heap)[2] for _ in range(min(top_n, len(heap)))]

    result = {
        "Top Popular in Same Cluster": top_cluster_recs
    }

    if use_location:
        # Query more neighbors to ensure enough after filtering
        dist, ind = balltree.query(user_coord, k=top_n + 10)  # distances in radians

        filtered_indices = [i for i in ind[0] if datarecords.iloc[i]['Destination_Name'] != user_destination]

        # Slice top_n after filtering
        filtered_indices = filtered_indices[:top_n]

        nearby_candidates = datarecords.iloc[filtered_indices]

        # Filter nearby destinations by type if specified
        if filter_type:
            nearby_candidates = nearby_candidates[nearby_candidates['Type'] == filter_type]



        if nearby_candidates.empty:
            result["Nearby Destinations"] = []
        else:
            # Calculate personalized scores mixing popularity and proximity
            earth_radius_km = 6371.0
            distances_km = dist[0][:len(nearby_candidates)] * earth_radius_km
            popularities = nearby_candidates['Popularity_Score'].values

            # Normalize distances and popularity to [0,1]
            norm_dist = (distances_km.max() - distances_km) / distances_km.max() if distances_km.max() != 0 else np.ones_like(distances_km)
            norm_pop = (popularities - popularities.min()) / (popularities.max() - popularities.min()) if popularities.max() != popularities.min() else np.ones_like(popularities)

            scores = weight_popularity * norm_pop + weight_proximity * norm_dist

            nearby_candidates = nearby_candidates.assign(Score=scores)
            nearby_candidates_sorted = nearby_candidates.sort_values(by='Score', ascending=False).head(top_n)

            result["Nearby Destinations"] = nearby_candidates_sorted.to_dict(orient='records')

    return result

# Testing the recommendation function:
user_dest = "Colosseum"
recs = recommend_travel(user_dest, top_n=5, use_location=True, filter_type=None)

print("Recommendations for " + user_dest + ":\n")

print("\nTop Popular Destinations in Same Cluster:")
for d in recs["Top Popular in Same Cluster"]:
    print("- " + d['Destination_Name'] + " (Popularity Score: " + str(d['Popularity_Score']) + ")")

print("\nNearby Destinations:")
nearby = recs.get("Nearby Destinations", [])
if not nearby:
    print("No nearby destinations found.")
else:
    for d in nearby:
        print("- " + d['Destination_Name'] + 
              " (Lat: " + str(d['Latitude']) + ", Lon: " + str(d['Longitude']) + 
              ", Popularity Score: " + str(d['Popularity_Score']) + ")")
```
